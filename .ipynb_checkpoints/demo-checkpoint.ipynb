## Data driven cricket analysis

Setting up dependencies

```python
%matplotlib inline

import pandas as pd
pd.set_option('precision', 2)
```

```python
batsman_innings = pd.read_pickle('summary_pickles/batsman_innings.pkl')
batsman_year = pd.read_pickle('summary_pickles/batsman_year.pkl')
bowler_spells = pd.read_pickle('summary_pickles/bowler_spells.pkl')
bowler_year = pd.read_pickle('summary_pickles/bowler_year.pkl')
bowling_overall = pd.read_pickle('summary_pickles/bowling_overall.pkl')
batting_overall = pd.read_pickle('summary_pickles/batting_overall.pkl')
chasing_under_pressure = pd.read_pickle('summary_pickles/chasing_under_pressure.pkl')
first_inning_bat = pd.read_pickle('summary_pickles/first_inning_bat.pkl')
second_inning_bat = pd.read_pickle('summary_pickles/second_inning_bat.pkl')
first_inning_first_6 = pd.read_pickle('summary_pickles/first_inning_first_6.pkl')
first_inning_middle_9 = pd.read_pickle('summary_pickles/first_inning_middle_9.pkl')
first_inning_last_4 = pd.read_pickle('summary_pickles/first_inning_last_4.pkl')
```

### Best batsmen batting first
